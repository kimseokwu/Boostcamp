# 조건부 확률이란?

- 베이즈 통계학을 이해하기 위해선 조건부 확률의 개념을 이해해야 합니다.

- 조건부확률 $P(A|B)$는 사건 B가 일어난 상황에서 사건 A가 발생할 확률을 의미한다.

$$
P(A \cap B) = P(B)P(A|B)
$$

 - 베이즈 정리는 조건부확률을 이용하여 정보를 갱신하는 방법을 알려줍니다.

- A라는 새로운 정보가 주어졌을때 P(B)로 부터 P(B|A)를 계산하는 방법을 제공한다.

$$
P(B|A) = \frac{P (A \cap B)}{P(A)} = P(B)\frac{P(A|B)}{P(A)}
$$


- $P(D): 데이터 \quad P(\theta): 가설$
- $P(\theta|D)$: 사후확률
- $P(\theta)$: 사전확률
- $P(D|\theta)$: 가능도(likelihood)
- $P(D)$: Evidence

$$
P(\theta|D) = P(\theta) \frac{P(D|\theta)}{P(D)}
$$

# 베이즈 정리: 예제

- COVID-99의 발병률이 10%로 알려져있다. COVID-99에 실제로 결렸을 때 , 검진될 확률은 99%, 실제로 걸리지 않았을 때 오검진될 확률이 1%라고 할 때, 어떤 사람이 질병에 걸렸다고 결과가 나왔을 때 정말로 COVID-99에 감염되었을 확률은?

- $P(\theta)=0.1$ (사전확률)
- $P(D|\theta)=0.99$ (가능도)
- $P(D|not \theta)=0.01$

- Evidence

$$
P(D) = \sum_{\theta}P(D|\theta)P(\theta) = 0.99 * 0.1 + 0.01 * (1 - 0.1) = 0.108
$$

- 따라서,

$$
P(\theta|D) = 0.1 * \frac{0.99}{0.108} = 0.916
$$

# 베이즈 정리를 통한 정보의 갱신

- 베이즈 정리를 통해 새로운 데이터가 들어왔을 때 앞서 계산한 사후확률을 사전확률로 사용하여 갱신된 사후확률을 계산할 수 있습니다.

- 앞서 COVID-99 판정을 받은 사람이 두 번째 검진을 받았을 때도 양성이 나왔을 때 진짜 COVID-99에 걸렸을 확률은?

- $P(D^*) = 0.99 * 0.524 + 0.1 * 0.476 = 0.566$

- $(갱신된 사후확률) = 0.524 * \frac{0.99}{0.566} = 0.917$

# 조건부 확률 = 인과관계?

- 조건부 확률은 유용한 통계적 해석을 제공하지만 인과관계(causality)를 추론할 때 함부로 사용해서는 안 됩니다. 데이터가 많아져도 조건부 확률만 가지고 인과관계를 추론하는 것은 불가능합니다.

- 인과관계는 데이터 분포의 변화에 강건한 예측모형을 만들 때 필요합니다.

- 조건부 확률에 기반한 예측모형은 유입되는 데이터의 분포가 바뀌었을 떄 예측확률이 크게 변할 수 있습니다.

- 인과관계를 알아내기 위해서는 두 변인에 모두 영향을 주는 중첩요인(confounding factor)의 효과를 제거하고 원인에 해당하는 변수만의 인과관계를 계산해야 합니다.

# 인과관계 추론: 예제

|   분류   |   Overall    | 신장결성 크기 작음 | 신장결석 크기 큼 |
| :------: | :----------: | :----------------: | :--------------: |
| 치료법 a | 78%(273/350) |     93%(81/87)     |   73%(192/273)   |
| 치료법 b | 83%(289/350) |    87%(234/270)    |    69%(55/80)    |

- T: 치료법, R: 완치, Z: 신장결석 크기
- 전체 데이터를 봤을 때는 치료법 B의 완치율이 더 높지만 신장결석 크기를 작을 때와 클 때로 나누면 두 경우 모두 치료법 A가 완치율이 높다.

- 신장결석 크기에 따른 중첩효과를 제거해야만 실제 치료법에 따른 완치율을 계산하는 것이 가능하다.

- 모든 환자가 치료법 A를 선택했을 때 완치율

$$
\frac{81}{87} * \frac{(87 + 270)}{700} + \frac{192}{263} * \frac{(263 + 80)}{700} = 0.8325
$$

- 모든 환자가 치료법 B를 선택했을 때 완치율

$$
\frac{234}{270} * \frac{(270 + 87)}{700} + \frac{55}{80} * \frac{(80 + 263)}{700} = 0.7789
$$
